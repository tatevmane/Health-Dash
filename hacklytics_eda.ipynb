```python
food = pd.read_csv("/Users/tatevgomtsyan/Independent Projects/FoodAccessResearchAtlasData2019.csv")
obesity = pd.read_csv("/Users/tatevgomtsyan/Downloads/LakeCounty_Health_-6821310239639457742.csv")
```

```python
merged_df = pd.merge(food, obesity, left_on='State',right_on='NAME')
merged_df = merged_df.drop(columns=['NAME', 'Shape__Area', 'Shape__Length'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

selected_variables = ['Obesity', 'PovertyRate', 'TractBlack', 'MedianFamilyIncome', 'LA1and20']  # Add other variables as needed
data_for_pairplot = merged_df[selected_variables].dropna()

# Create pair plot
sns.pairplot(data_for_pairplot)

# Show plot
plt.show()
```

The above pair plot shows the relationship between various variables of interest in the dataset. Noteworthy points: As poverty rate increases, median family income dereases. Obesity and poverty seem to be slightly positively correlated, although there is no clear relationship. 

```python
# Generate summary statistics for the obesity variable
obesity_summary = merged_df['Obesity'].describe()

# Print the summary table
print(obesity_summary)
```

```python
# Sort DataFrame by 'Obesity' column in descending order
sorted_df = merged_df.sort_values(by='Obesity', ascending=False)

# Create bar plot of obesity by state
plt.figure(figsize=(12, 6))  # Set the figure size
sns.barplot(data=sorted_df, x='State', y='Obesity', palette='viridis')

# Set plot title and labels
plt.title('Bar Plot of Obesity by State')
plt.xlabel('State')
plt.ylabel('Obesity Rate')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()  # Adjust layout
plt.show()
```

Southern states like Louisiana, Mississippi, and Alabama have the highest obesity rates in the country, while more metropolitan or nature-surrounded cities have lower rates. 

```python
# Generate summary statistics for the obesity variable
obesity_summary = merged_df['PovertyRate'].describe()

# Print the summary table
print(obesity_summary)
```

```python
# Sort DataFrame by 'PovertyRate' column in descending order
poverty = merged_df.sort_values(by='PovertyRate', ascending=False)

# Create bar plot of poverty rate by state
plt.figure(figsize=(12, 6))  # Set the figure size
sns.barplot(data=poverty, x='State', y='PovertyRate')

# Set plot title and labels
plt.title('Bar Plot of Poverty Rate by State')
plt.xlabel('State')
plt.ylabel('Poverty Rate')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()  # Adjust layout
plt.show()
```

Indiana has the highest poverty rate, followed by Maryland.

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=merged_df, x='Urban', y='Obesity', palette='coolwarm')

# Set plot title and labels
plt.title('Obesity Rate by Urban/Rural Status')
plt.xlabel('Rural/Urban')
plt.ylabel('Obesity Rate')

# Show plot
plt.tight_layout()
plt.show()
```

Obesity rate is not visibly much different in rural vs urban areas. Obesity rate across the US sits around 30%, making it a big public health issue.
